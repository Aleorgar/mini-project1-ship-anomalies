A low lubrication oil temperature may indicate it is not reaching its optimal operating temperature, potentially causing inadequate lubrication.
- **Coolant temperature:** High coolant temperature signals overheating, which various issues, including a failed thermostat, coolant leak, or insufficient coolant flow can cause. A low coolant temperature could suggest the engine is not reaching its optimal operating temperature, affecting performance and efficiency.

Issues with engines could lead to engine malfunctions, potential safety hazards, and downtime (e.g. delayed deliveries), resulting in the breakdown of a ship’s overall functionality, consequently impacting the business, such as affecting revenue via failure to deliver goods. ##1. Setup and Data Import

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.decomposition import PCA

# Load dataset
url = 'https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/engine.csv'
data = pd.read_csv(url)
data.head()
```

##3. Outlier Detection with IQR Method

```python
# IQR Outlier Detection
iqr_outliers = pd.DataFrame()
for col in data.columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    iqr_outliers[col + '_outlier'] = ((data[col] < lower_bound) | (data[col] > upper_bound)).astype(int)

# Count outliers per row
iqr_outliers['total_outliers'] = iqr_outliers.sum(axis=1)
data['IQR_anomaly'] = (iqr_outliers['total_outliers'] >= 2).astype(int)

# Percentage of anomalies
outlier_percent = data['IQR_anomaly'].mean() * 100
print(f'IQR-based anomalies: {outlier_percent:.2f}%')
```

The Interquartile Range (IQR) method is a simple yet effective statistical technique to detect outliers by flagging values that lie beyond 1.5 times the IQR from the lower or upper quartile. This approach is straightforward and does not require assumptions about the data distribution, making it useful for preliminary outlier detection.

In our case, we flagged each feature independently and considered a sample an outlier if at least two or more features were outside the IQR range. This threshold helped us narrow the anomalies down to approximately 2.16% of the samples, which aligns with the expected anomaly proportion (1–5%).

While IQR is useful for identifying extreme values, it can miss multivariate anomalies that only become apparent when multiple features are analyzed together. It also assumes that outliers are extreme in magnitude, which may not always be the case in more complex datasets. Therefore, while IQR provides a strong baseline, more advanced ML-based methods are necessary for capturing subtle or contextual anomalies.

```python
# Remove previously generated _flagged columns (if any)
data = data.copy()
data = data[[col for col in data.columns if not col.endswith('_flagged')]]

# Create new _flagged columns using IQR method
for col in data.select_dtypes(include='number').columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data.loc[:, f'{col}_flagged'] = ((data[col] < lower) | (data[col] > upper)).astype(int)

# Display only original features and _flagged columns (excluding IQR_anomaly if present)
columns_to_display = [col for col in data.columns if '_anomaly' not in col]
data[columns_to_display].head()
```

```python
print(data.columns.tolist())

# Create IQR-flagged columns (0 or 1) for each numeric feature
columns_to_check = ['Engine rpm', 'Lub oil pressure', 'Fuel pressure',
                    'Coolant pressure', 'lub oil temp', 'Coolant temp']  # Use exact column names

for col in columns_to_check:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data[f'{col}_flagged'] = ((data[col] < lower) | (data[col] > upper)).astype(int)

# Count how many features are flagged as outliers
flagged_cols = [f'{col}_flagged' for col in columns_to_check]
data['IQR_preds'] = (data[flagged_cols].sum(axis=1) >= 2).astype(int)

# Display the result
data[flagged_cols + ['IQR_preds']].head()
```

## 4. Feature Scaling (only for SVM, Isolation and PCA)

###4.1. One-Class SVM & PCA

```python
from sklearn.svm import OneClassSVM
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Scale the data
cols_to_scale = ['Engine rpm', 'Lub oil pressure', 'Fuel pressure',
                 'Coolant pressure', 'lub oil temp', 'Coolant temp']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data[cols_to_scale])

# Apply One-Class SVM model
svm = OneClassSVM(kernel='rbf', gamma=0.5, nu=0.045)  # 4.94% de anomalías
svm.fit(X_scaled)
svm_preds = svm.predict(X_scaled)

# Store mapped predictions in the DataFrame
data['SVM_anomaly'] = pd.Series(svm_preds, index=data.index).map({1: 0, -1: 1})

# Show array breakdown
print("Predictions array breakdown:")
print(np.unique(svm_preds, return_counts=True))

# Show porcentage
svm_percent = data['SVM_anomaly'].mean() * 100
print(f"\nSVM-based anomalies: {svm_percent:.2f}%")

# Full table
tabla_completa = data[['Engine rpm', 'Lub oil pressure', 'Fuel pressure',
                       'Coolant pressure', 'lub oil temp', 'Coolant temp',
                       'SVM_anomaly']]
display(tabla_completa.head())

# PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(pca_components, columns=['PC1', 'PC2'], index=data.index)
pca_df['anomaly'] = data['SVM_anomaly']
display(pca_df.head())

# Final plot with correct colors
labels = {0: 'Normal observations', 1: 'Anomaly'}
colors = {0: 'deepskyblue', 1: 'red'}

plt.figure(figsize=(10, 8))
for label in [0, 1]:
    subset = pca_df[pca_df['anomaly'] == label]
    plt.scatter(subset['PC1'], subset['PC2'],
                c=colors[label], label=labels[label],
                alpha=0.5, edgecolor='k')

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('One-Class SVM Anomaly Detection in PCA Space')
plt.legend()
plt.show()
```

###4.2. Isolation Forest & PCA (ML-based Anomaly Detection)

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Columns to use
cols_to_use = ['Engine rpm', 'Lub oil pressure', 'Fuel pressure',
               'Coolant pressure', 'lub oil temp', 'Coolant temp']

# Scale data ONLY for PCA (NOT for Isolation Forest)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data[cols_to_use])

# Train Isolation Forest on UNscaled data
iso = IsolationForest(contamination=0.05, random_state=42)
iso_preds = iso.fit_predict(data[cols_to_use])
data['IF_anomaly'] = pd.Series(iso_preds, index=data.index).map({1: 0, -1: 1})

# Show array breakdown
print("Isolation Forest predictions breakdown:")
pred_breakdown = np.unique(iso_preds, return_counts=True)
print(pred_breakdown)

# Anomaly's percentage
iso_percent = data['IF_anomaly'].mean() * 100
print(f"\nIsolation Forest anomalies: {iso_percent:.2f}%")

# Full table with variables + anomaly
tabla_completa = data[cols_to_use + ['IF_anomaly']]
print("\nTabla completa con variables + IF_anomaly:")
display(tabla_completa.head())

# PCA on SCALED data
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)

pca_df_IF = pd.DataFrame(pca_components, columns=['PC1', 'PC2'], index=data.index)
pca_df_IF['anomaly'] = data['IF_anomaly']
print("\nTabla PCA con PC1, PC2 y anomaly:")
display(pca_df_IF.head())

# Visualization
labels = {0: 'Normal observations', 1: 'Anomaly'}
colors = {0: 'deepskyblue', 1: 'red'}

plt.figure(figsize=(10, 8))
for label in [0, 1]:
    subset = pca_df_IF[pca_df_IF['anomaly'] == label]
    plt.scatter(subset['PC1'], subset['PC2'],
                c=colors[label], label=labels[label],
                alpha=0.5, edgecolor='k')

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('Isolation Forest Anomaly Detection in PCA Space')
plt.legend()
plt.show()
```
